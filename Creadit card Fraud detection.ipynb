```python
#according to heatmap y(class) is dependent to all the x (V1,V2,V2...)
X = data.drop(['Class'],axis = 1)
Y = data['Class']
print (X.shape)
print (Y.shape)

xdata =X.values 
ydata =Y.values
```

```python
#by using scikit learn to split data into training and testing test sets
from sklearn.model_selection import train_test_split
#split the data into TRAINING AND  TESTING SETS 


xTrain,xTest,yTrain,yTest = train_test_split(
     xdata,ydata ,test_size= 0.2 ,random_state = 42)
```

```python
#building the Random forest classifier (Random forest)
from sklearn.ensemble import RandomForestClassifier
#RandomForest model 
RFC = RandomForestClassifier()
RFC.fit(xTrain, yTrain)
#prediction
ypred = RFC.prediction(xTest)
```

```python
#evaluating the classifier 
#printingn score 
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import f1_score,matthews_corrcoef
from sklearn.metrics import confusion_matrix

n_outliers = len(fraud)
n_errors = (ypred != yTest).sum()
print("The model used is RandomForest Classifier")

```

```python
acc= accuracy_score(yTest,ypred)
print ("The accuracy is {}".format(acc))

```

```python
prec = prescision_score(yTest,ypred)
print( "The precision is {}".format(prec))
```

```python
rec = recall_score(yTest,ypred)
print ("The recall_score is {}".format(rec))
```

```python
f1 = f1_score(yTest,ypred)
print ("The f1_score is {}".format(f1))
```

```python
mcc = matthews_corrcoef(yTest,ypred)
print( "The Matthews_corellation coefficient is {}".format(mcc))
```

```python
#printing the confusion matrixs 
LABELS = ['Normal','Fraud']
conf_matrix = confusion_matrix(yTest,ypred)
plt.figure(figsize= (11,11))
sns.heatmap(conf_matrix, xticklabels = LABELS,
                        yticklabels  = LABELS, annot = True , fmt ="d");
plt.title ("Confusion_Matrix")
plt.ylabel("True class")
plt.xlabel("predicted class")
plt.show()
```
